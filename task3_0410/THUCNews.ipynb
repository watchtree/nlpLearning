```python
def stopwordslist():
    stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords
```

```python
#取前100条数据加快处理
data = lists[:100]
```

```python
#统计词表出现次数
wordList = {}
for i in data:
    for j in i:
        if j in wordList:
            wordList[j] += 1
        else:
            wordList[j] = 1
#创建词表
count = 0
dic = {}
for i, j in wordList.items():
    if j > 2:
        dic[i] = count 
        count += 1
#反向词表
resDic = {}
for i, j in dic.items():
    resDic[j] = i
```
